## Defino a técnica de criação de amostras sintéticas para balanceamento.

```python
# Define SMOTE-Tomek Links
sm = SMOTE(
    sampling_strategy='auto',
    random_state=11,
    k_neighbors=5,
    n_jobs=4
)

tl = TomekLinks(
    sampling_strategy='all',
    n_jobs=4)

over=SMOTETomek(
    sampling_strategy='auto',
    random_state=12,
    smote=sm,
    tomek=tl,
    n_jobs=4
)
```

## Definição de alguns parâmetros para a pipeline e o modelo.

```python
skb = SelectKBest(score_func=f_classif, k=10)
rfe = RFE(estimator=LogisticRegression(),n_features_to_select=5)
var_thr = VarianceThreshold(threshold = 0.01)
params_mlp = {
    'combined_features__skb__k':[3, 5, 7, 10, 15, 20],
    'combined_features__rfe__n_features_to_select':[3, 5, 7],
    'classifier__hidden_layer_sizes': [(200,50, 30), (100,50, 10),(100,50),(200,100),(500,250), (20,), (50,), (100,), (10,), (200,)],
    'classifier__activation': ['tanh', 'relu'],
    'classifier__solver': ['sgd', 'adam'],
    'classifier__alpha': [0.0001, 0.005, 0.05],
    'classifier__learning_rate': ['constant', 'adaptive']
}

```

###Combine features selection

```python
X = (dados.drop(columns=['Obito_MS'])).values
y = (dados['Obito_MS']).values
```

```python
combined_features = FeatureUnion([("skb", skb), ("rfe", rfe)])

X_features = combined_features.fit(X, y).transform(X)
print("Combined space has", X_features.shape[1], "features")
```

##Pipeline com passo a passo a ser executado.

```python
def get_pipeline(cls):
    return Pipeline([('scaler', MinMaxScaler()),
                     ('var', var_thr),
                     ('combined_features', combined_features),
                     ('classifier', cls)])
```

###Escolha dos melhores hiperparâmetros e treinamento e avalição do modelo

```python
metrics_dict = {
    'acc_train': [],
    'f1_train': [],
    'recall_train': [],
    'precision_train': [],
    'roc_auc_train': [],
    'acc_test': [],
    'f1_test': [],
    'recall_test': [],
    'precision_test': [],
    'roc_auc_test': []
}


# Realização do Hold-out
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=12, stratify=y)

# Aplicação do SMOTE
X_train_smt, y_train_smt = over.fit_resample(X_train, y_train)

mlp = get_pipeline(MLPClassifier(random_state=12))

# Treinamento do modelo MLP com GridSearchCV
gs_mlp = GridSearchCV(mlp, param_grid=params_mlp, scoring='f1', cv=10).fit(X_train_smt, y_train_smt)
modelmlp = gs_mlp.best_estimator_

n_folds = 5
skfold = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=12)

tpr_list = []
mean_fpr = np.linspace(0, 1, 100)


for fold, (train_index, test_index) in enumerate(skfold.split(X, y)):

    X_train, X_test = [X[i] for i in train_index], [X[i] for i in test_index]
    y_train, y_test = [y[i] for i in train_index], [y[i] for i in test_index]


    X_train_smt, y_train_smt = over.fit_resample(X_train, y_train)
    y_pred_train_mlp = np.round(modelmlp.predict(X_train_smt))

    metrics_dict['acc_train'].append(accuracy_score(y_train_smt, y_pred_train_mlp))
    metrics_dict['f1_train'].append(f1_score(y_train_smt, y_pred_train_mlp))
    metrics_dict['recall_train'].append(recall_score(y_train_smt, y_pred_train_mlp))
    metrics_dict['precision_train'].append(precision_score(y_train_smt, y_pred_train_mlp))
    metrics_dict['roc_auc_train'].append(roc_auc_score(y_train_smt, y_pred_train_mlp))

    # Métricas para teste
    y_pred_mlp = modelmlp.predict(X_test)

    metrics_dict['acc_test'].append(accuracy_score(y_test, y_pred_mlp))
    metrics_dict['f1_test'].append(f1_score(y_test, y_pred_mlp))
    metrics_dict['recall_test'].append(recall_score(y_test, y_pred_mlp))
    metrics_dict['precision_test'].append(precision_score(y_test, y_pred_mlp))
    metrics_dict['roc_auc_test'].append(roc_auc_score(y_test, y_pred_mlp))

    y_prob = modelmlp.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    tpr_list.append(np.interp(mean_fpr, fpr, tpr))
```

##Criar e mostrar a tabela de métricas resultantes do modelo.Assim como os gráficos de curva

```python
def calcular_media_individual_por_chave(dicionario):
    medias_por_chave = {}
    std_por_chave = {}

    for chave, valores in dicionario.items():
            media = np.mean(valores)
            std = np.std(valores)
            medias_por_chave[chave] = media
            std_por_chave[chave] = std
    return medias_por_chave, std_por_chave


medias_individual_por_chave = calcular_media_individual_por_chave(metrics_dict)
```
